# Linear Regression (Python) — AI-assisted Assignment 3

This notebook models **Salary** from **YearsExperience** using ordinary least squares linear regression.

We report:
- slope and y-intercept of the fitted line
- Pearson correlation coefficient (*r*)
- mean squared error (MSE)

and display a scatter plot with the fitted line and text annotations for the equation and *r*.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error

dataset = pd.read_csv("regression_data-1.csv")
dataset.head()
```

## What is being calculated

We fit a straight line:

$$\text{Salary} \approx b + m \cdot \text{YearsExperience}$$

| Quantity | Meaning |
|----------|---------|
| **Slope (*m*)** | Change in predicted Salary for each additional year of experience |
| **Intercept (*b*)** | Predicted Salary when YearsExperience = 0 |
| **Correlation (*r*)** | Strength and direction of the linear association (−1 to 1) |
| **MSE** | Average squared difference between observed and predicted Salary |

```python
x = dataset["YearsExperience"]
y = dataset["Salary"]

slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_pred = slope * x + intercept
mse = mean_squared_error(y, y_pred)

print("Slope:", slope)
print("Intercept:", intercept)
print("Correlation coefficient (r):", r_value)
print("Mean Squared Error (MSE):", mse)
```

## Interpretation of the results

- **Slope:** For each additional year of experience, predicted salary increases by about **$8,285**.
- **Intercept:** When experience is 0 years, the model predicts a starting salary of about **$29,204**. This is an extrapolation and may or may not be realistic for the hiring context.
- **Correlation (*r* ≈ 0.89):** There is a **strong positive** linear relationship: higher experience tends to go with higher salary.
- **MSE ≈ 1.75×10⁷:** The average squared prediction error. The square root (RMSE) is about **$4,186**, so typical prediction misses are a few thousand dollars on this dataset.

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color="red", label="Original data")
plt.plot(x, y_pred, color="blue", label="Fitted line")
plt.text(
    min(x) + (max(x) - min(x)) * 0.55,
    min(y) + (max(y) - min(y)) * 0.15,
    f"y = {slope:.2f}x + {intercept:.2f}\n"
    f"r = {r_value:.2f}",
    fontsize=12,
)
plt.title("Salary vs YearsExperience")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.legend()
plt.tight_layout()
plt.show()
```
