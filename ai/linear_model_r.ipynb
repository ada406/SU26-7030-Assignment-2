# Linear Regression (R) — AI-assisted Assignment 3

This notebook models **Salary** from **YearsExperience** using ordinary least squares linear regression in R.

We report slope, intercept, Pearson's correlation (*r*), and mean squared error (MSE), and annotate the plot with the fitted equation and *r*.

```r
dataset <- read.csv("regression_data-1.csv")
head(dataset)
```

## What is being calculated

We fit:

`Salary ≈ intercept + slope × YearsExperience`

- **Slope:** change in predicted Salary for each additional year of experience
- **Intercept:** predicted Salary when YearsExperience = 0
- **Correlation (*r*):** strength/direction of the linear relationship
- **MSE:** average squared difference between observed and predicted Salary

```r
model <- lm(Salary ~ YearsExperience, data = dataset)
slope <- coef(model)[2]
intercept <- coef(model)[1]
r <- cor(dataset$YearsExperience, dataset$Salary)
pred <- predict(model)
mse <- mean((dataset$Salary - pred)^2)

cat("Slope:", slope, "\n")
cat("Intercept:", intercept, "\n")
cat("Correlation coefficient (r):", r, "\n")
cat("Mean Squared Error (MSE):", mse, "\n")
```

## Interpretation of the results

- **Slope:** For each additional year of experience, predicted salary increases by about **$8,285**.
- **Intercept:** At 0 years of experience, the model predicts about **$29,204**. Treat this carefully if zero experience is outside the useful range of the data.
- **Correlation (*r* ≈ 0.89):** Strong positive linear association between experience and salary.
- **MSE ≈ 1.75×10⁷:** Average squared error; RMSE ≈ **$4,186**, so predictions are typically off by a few thousand dollars on this sample.

```r
library(ggplot2)

label_text <- paste(
  "y =", round(slope, 2), "x +", round(intercept, 2),
  "\nr =", round(r, 2)
)

x_pos <- min(dataset$YearsExperience) +
  (max(dataset$YearsExperience) - min(dataset$YearsExperience)) * 0.55
y_pos <- min(dataset$Salary) +
  (max(dataset$Salary) - min(dataset$Salary)) * 0.15

ggplot(dataset, aes(x = YearsExperience, y = Salary)) +
  geom_point(color = "red") +
  geom_smooth(method = "lm", se = FALSE, color = "blue") +
  annotate("text", x = x_pos, y = y_pos, label = label_text, hjust = 0, size = 4) +
  ggtitle("Salary vs YearsExperience") +
  xlab("YearsExperience") +
  ylab("Salary")
```
